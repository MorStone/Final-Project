```python
Purchases=df["Rev_D7"][(df["Opt_Goal"] == "Purchases")]
Value=df["Rev_D7"][(df["Opt_Goal"] == "Value")]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=plt.figaspect(1))
values = [Purchases.mean(),Value.mean()] 
labels = ['Purchases','Value']
plt.title('Average Rev_D7 per Opt_Goal')
colors = ['b','g']
explode = (0, 0.1)

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


plt.pie(values,colors=colors,explode=explode, labels=labels, autopct=make_autopct(values),shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
stats.ttest_ind(Purchases, Value)
```

```python
Tier1=df["Rev_D7"][(df["GEO"] == "Tier 1")]
US=df["Rev_D7"][(df["GEO"] == "US")]
Worldwide=df["Rev_D7"][(df["GEO"] == "Worldwide")]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=plt.figaspect(1))
values = [Tier1.mean(),US.mean(),Worldwide.mean()] 
labels = ['Tier1','US','Worldwide']
plt.title('Average Rev_D7 per GEO')
colors = ['b','g','y']
explode = (0.02, 0.02,0.02)

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


plt.pie(values,colors=colors,explode=explode, labels=labels, autopct=make_autopct(values),shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
n_groups = 3
S=(Tier1.mean(), 0,0 )
D=( 0, US.mean(),0  )
V=(0, 0, Worldwide.mean())

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.3
opacity = 0.9
 
rects1 = plt.bar(index, S, bar_width,
                 alpha=opacity,
                 color='g')

rects2 = plt.bar(index, D, bar_width,
                 alpha=opacity,
                 color='b')

rects3 = plt.bar(index, V, bar_width,
                 alpha=opacity,
                 color='r')

 
plt.xlabel('GEO')
plt.ylabel('Mean_Rev_D7')
plt.title('Average Rev_D7 per GEO')
plt.xticks(index,('Tier1','US','Worldwide'))
plt.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.00*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
plt.tight_layout()
plt.show()
```

```python
df_temp=df
```

```python
df_temp['Opt_Goal_Code']=0

for i in range(len(df)):
    if (df_temp['Opt_Goal'][i]=="Purchases"):
        df_temp['Opt_Goal_Code'][i]=1
    elif (df_temp['Opt_Goal'][i]=="Value"):
        df_temp['Opt_Goal_Code'][i]=2
```

```python
df_temp['Device_Code']=0

for i in range(len(df)):
    if (df_temp['Device'][i]=="iPad"):
        df_temp['Device_Code'][i]=1
    elif (df_temp['Device'][i]=="Android"):
        df_temp['Device_Code'][i]=2
    elif (df_temp['Device'][i]=="iPhone"):
           df_temp['Device_Code'][i]=3
```

```python
import numpy as np
from sklearn.utils import shuffle
A=df_temp
A = shuffle(A)
```

```python
test=A[['Installs_D3','Rev_D3','Opt_Goal_Code','Device_Code','FTD_D3']]
target=A['Rev_D7_target']
test_ins=A[['Impressions_D3','Clicks_D3','Installs_D3']]
target_ins=A['Installs_D7_target']
```

```python
train=test[['Opt_Goal_Code','Device_Code','Installs_D3','Rev_D3']]
train_ins=test_ins[['Impressions_D3','Clicks_D3','Installs_D3']]
```

```python
from sklearn import neighbors 
knn = neighbors.KNeighborsClassifier(3) 
knn.fit(train[:100], target[:100])
knn.score(train[100:], target[100:])
```

```python
from sklearn import neighbors 
knn = neighbors.KNeighborsClassifier(3)
knn.fit(train_ins[:100], target_ins[:100])
knn.score(train_ins[100:], target_ins[100:])
```

```python
from sklearn import svm 
svc = svm.SVC(kernel='linear') 
svc.fit(train[:100], target[:100]) 
svc.score(train[100:], target[100:])
```
